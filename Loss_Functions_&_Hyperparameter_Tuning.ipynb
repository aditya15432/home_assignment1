```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define true values (one-hot encoded) and predictions
y_true = tf.constant([[0., 1., 0.]], dtype=tf.float32)

# Two sets of predictions for comparison
y_pred_1 = tf.constant([[0.1, 0.8, 0.1]], dtype=tf.float32)  # Closer prediction
y_pred_2 = tf.constant([[0.3, 0.4, 0.3]], dtype=tf.float32)  # Less accurate

# Step 2: Compute MSE and Categorical Cross-Entropy losses
mse_1 = tf.keras.losses.MeanSquaredError()(y_true, y_pred_1).numpy()
mse_2 = tf.keras.losses.MeanSquaredError()(y_true, y_pred_2).numpy()

cce_1 = tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred_1).numpy()
cce_2 = tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred_2).numpy()

# Step 3: Print loss values
print("Loss Comparison:")
print(f"MSE (Closer prediction): {mse_1:.4f}")
print(f"MSE (Less accurate):     {mse_2:.4f}")
print(f"CCE (Closer prediction): {cce_1:.4f}")
print(f"CCE (Less accurate):     {cce_2:.4f}")

# Step 4: Plot loss values using bar chart
labels = ['MSE', 'CCE']
closer = [mse_1, cce_1]
farther = [mse_2, cce_2]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, closer, width, label='Closer Prediction', color='green')
plt.bar(x + width/2, farther, width, label='Less Accurate Prediction', color='red')

plt.ylabel('Loss Value')
plt.title('Comparison of MSE and Categorical Cross-Entropy')
plt.xticks(x, labels)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
